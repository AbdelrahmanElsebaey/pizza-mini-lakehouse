##Define Ingestion Configuration

```python
INGESTION_CONFIG=[{
    "source":"pizzaplace",
    "path": "/Volumes/workspace/bronze/raw_sources/pizzaplace/order_details.csv",
    "table": "order_details"
},{
    "source":"pizzaplace",
    "path": "/Volumes/workspace/bronze/raw_sources/pizzaplace/orders.csv",
    "table": "orders"
},{
    "source":"pizzaplace",
    "path":"/Volumes/workspace/bronze/raw_sources/pizzaplace/pizza_types.csv",
    "table": "pizza_types"
},{
    "source":"pizzaplace",
    "path": "/Volumes/workspace/bronze/raw_sources/pizzaplace/pizzas.csv",
    "table": "pizzas"
}]
```

##Ingest Files into Bronze Tables

```python
#reading and writing data
for i in INGESTION_CONFIG:
    df=(spark.read.option("header",True).option("inferSchema",True).csv(i["path"]))
    df.write.mode("overwrite").format("delta").saveAsTable(f"workspace.bronze.{i['table']}")
    print(f"Table {i['table']} is Created succesfully!")
```
