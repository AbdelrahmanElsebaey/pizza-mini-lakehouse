```python
query="""
SELECT
  od.order_details_id as fact_id,
  od.pizza_id,
  od.quantity,
  o.date AS order_date,
  o.time AS order_time
FROM workspace.silver.orders AS o
LEFT JOIN workspace.silver.order_details AS od
ON o.order_id = od.order_id;"""
df=spark.sql(query)
```

```python
df.limit(5).display()
```

##saving

```python
df.write.mode('overwrite').format('delta').saveAsTable('workspace.gold.fact_orders')
print('Saved successfully!')
```
