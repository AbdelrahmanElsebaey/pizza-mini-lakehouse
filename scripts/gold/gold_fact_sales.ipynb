```python
query="""
SELECT
  od.order_id,
  od.pizza_id,
  od.quantity,
  p.price,
  p.price * od.quantity AS revenue
FROM workspace.silver.order_details AS od
LEFT JOIN workspace.silver.pizzas AS p
ON od.pizza_id = p.pizza_id; """
df=spark.sql(query)
```

```python
df.limit(5).display()
```

##saving

```python
df.write.mode("overwrite").format('delta').saveAsTable('workspace.gold.fact_sales')
print('Saved Successfully!')
```
