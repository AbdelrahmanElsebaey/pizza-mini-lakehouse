```python
query="""
SELECT DISTINCT
  o.date as order_date,
  YEAR(o.date) as year,
  MONTH(o.date) as month,
  DAY(o.date) as day,
  QUARTER(o.date) as quarter
FROM workspace.silver.orders as o"""

df=spark.sql(query)
```

```python
df.limit(5).display()
```

##saving

```python
df.write.mode('overwrite').format('delta').saveAsTable('workspace.gold.dim_date')
print('Saved successfully!')
```
