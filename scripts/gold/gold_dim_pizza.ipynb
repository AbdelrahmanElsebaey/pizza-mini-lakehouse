```python
query="""
SELECT 
  p.pizza_id,
  p.size,
  p.price,
  pt.name AS pizza_name,
  pt.category,
  pt.ingredients
FROM workspace.silver.pizzas as p
LEFT JOIN workspace.silver.pizza_types AS pt
ON p.pizza_type_id = pt.pizza_type_id;"""
df=spark.sql(query)
```

```python
#trying it
df.limit(10).display()
```

##saving 

```python
df.write.mode('overwrite').format('delta').saveAsTable('workspace.gold.dim_pizza')
print('Saved successfully!')
```
